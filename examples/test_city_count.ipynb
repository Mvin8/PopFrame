```python
import pandas as pd
import geopandas as gpd
from shapely.geometry import Point

# Определение функции для поиска и удаления дубликатов
def find_and_remove_duplicates(gdf, name_column, geometry_column, population_column):
    tolerance_km = 40
    tolerance_deg = tolerance_km / 111  # Преобразование километров в градусы (приблизительно)
    
    # Создание пространственного индекса
    gdf = gdf.reset_index(drop=True)
    gdf['duplicate'] = False
    
    spatial_index = gdf.sindex
    
    for idx, row in gdf.iterrows():
        if gdf.at[idx, 'duplicate']:
            continue
        
        # Поиск ближайших точек в пределах допустимого расстояния
        possible_matches_index = list(spatial_index.intersection(row[geometry_column].buffer(tolerance_deg).bounds))
        possible_matches = gdf.iloc[possible_matches_index]
        exact_matches = possible_matches[possible_matches[name_column] == row[name_column]]
        exact_matches = exact_matches[exact_matches[geometry_column].distance(row[geometry_column]) < tolerance_deg]
        
        # Проверка разницы в населении
        exact_matches = exact_matches[exact_matches[population_column].apply(lambda x: x <= row[population_column] * 1.5 and x >= row[population_column] / 1.5)]
        
        if len(exact_matches) > 1:
            duplicate_indices = exact_matches.index
            # Помечаем все, кроме одного, как дубликаты
            gdf.loc[duplicate_indices[1:], 'duplicate'] = True

    return gdf
```

```python
# Применение функции для поиска и удаления дубликатов
gdf_with_duplicates = find_and_remove_duplicates(towns, 'name', 'geometry', 'population')

# Удаление дубликатов из исходного GeoDataFrame
towns = gdf_with_duplicates[~gdf_with_duplicates['duplicate']].drop(columns=['duplicate'])

towns.reset_index(drop=True, inplace=True)
towns
```

```python
import pandas as pd
from geopandas import GeoDataFrame

def fill_levels(towns: GeoDataFrame, population_thresholds: dict[str, tuple[int, int]]) -> GeoDataFrame:
    def _assign_level(row, population_thresholds) -> str:
        population = row['population']
        for level, (lower_bound, upper_bound) in population_thresholds.items():
            if lower_bound < population <= upper_bound:
                return level
        return "Неизвестный уровень"  # На случай, если население не вписывается ни в один из диапазонов

    towns = towns.copy()
    towns["level"] = towns.apply(_assign_level, axis=1, args=(population_thresholds,))
    return towns

# Пример использования:
population_thresholds = {
    "Сверхкрупный город": (3000000, float('inf')),
    "Крупнейший город": (1000000, 3000000),
    "Крупный город": (250000, 1000000),
    "Большой город": (100000, 250000),
    "Средний город": (50000, 100000),
    "Малый город": (5000, 50000),
    "Крупное сельское поселение": (3000, 5000),
    "Большое сельское поселение": (1000, 3000),
    "Среднее сельское поселение": (200, 1000),
    "Малое сельское поселение": (0, 200),
}

# Заполнение уровней
updated_towns = fill_levels(towns, population_thresholds)
updated_towns
```

```python
# Проверка столбца 'level' на NaN или 0
nan_count = updated_towns['level'].isna().sum()
zero_count = (updated_towns['level'] == 0).sum()

# Вывод количества значений NaN и нулей
print(f"Количество NaN значений: {nan_count}")
print(f"Количество нулевых значений: {zero_count}")
```

```python
import pandas as pd
import geopandas as gpd

# Ваши данные level_mapping
level_mapping = {
    "Сверхкрупный город": "1.2.1",
    "Крупнейший город": "1.2.2",
    "Крупный город": "1.2.3",
    "Большой город": "1.2.4",
    "Средний город": "1.2.5",
    "Малый город": "1.2.6",
    "Крупное сельское поселение": "1.2.7",
    "Большое сельское поселение": "1.2.8",
    "Среднее сельское поселение": "1.2.9",
    "Малое сельское поселение": "1.2.10"
}

# Присвоение каждому городу района, в котором он находится
city_gdf = gpd.sjoin(updated_towns, district, how='left', op='within')

# Удаление строк с пустыми значениями в колонке 'name_right'
city_gdf = city_gdf.dropna(subset=['name_right'])

# Подсчет количества городов каждого уровня в каждом районе
level_counts = city_gdf.groupby(['name_right', 'level']).size().unstack(fill_value=0).stack().reset_index(name='count')

# Создание пустого DataFrame с нулями для всех комбинаций районов и уровней урбанизации
districts = district['name'].unique()
levels = list(level_mapping.keys())
all_combinations = pd.MultiIndex.from_product([districts, levels], names=['name_right', 'level']).to_frame(index=False)
all_combinations = all_combinations.merge(level_counts, how='left', on=['name_right', 'level']).fillna(0)


# Словарь для отображения названий показателей
level_names_mapping = {
    "Сверхкрупный город": "Количество сверхкрупных городов на территории",
    "Крупнейший город": "Количество крупнейших городов на территории",
    "Крупный город": "Количество крупных городов на территории",
    "Большой город": "Количество больших городов на территории",
    "Средний город": "Количество средних городов на территории",
    "Малый город": "Количество малых городов на территории",
    "Крупное сельское поселение": "Количество крупных сельских поселений на территории",
    "Большое сельское поселение": "Количество больших сельских поселений на территории",
    "Среднее сельское поселение": "Количество средних сельских поселений на территории",
    "Малое сельское поселение": "Количество малых сельских поселений на территории"
}

# Создание результирующего DataFrame
results = []

for _, row in all_combinations.iterrows():
    district = row['name_right']
    level = row['level']
    count = int(row['count'])
    key = level_mapping[level]
    name = level_names_mapping[level]
    results.append({
        '№ п/п': key,
        'Название хранимое': name,
        'ед.изм.': 'ед',
        'Значение': count,
        'Территория': district,
        'Источник': 'modeled',
        'Период': 2024
    })

results_gdf = pd.DataFrame(results).sort_values(by=['Территория', '№ п/п'])

# Добавление агрегированных данных
aggregate_territory_name = "Ленинградская область"
aggregate_data = results_gdf.groupby('Название хранимое').agg({'Значение': 'sum'}).reset_index()
aggregate_data['№ п/п'] = aggregate_data['Название хранимое'].map({v: level_mapping[k] for k, v in level_names_mapping.items()})
aggregate_data['ед.изм.'] = 'ед'
aggregate_data['Территория'] = aggregate_territory_name
aggregate_data['Источник'] = 'modeled'
aggregate_data['Период'] = 2024

# Приведение агрегированных данных в нужный порядок столбцов
aggregate_data = aggregate_data[['№ п/п', 'Название хранимое', 'ед.изм.', 'Значение', 'Территория', 'Источник', 'Период']]

# Добавление агрегированных данных в начало DataFrame
results_gdf = pd.concat([aggregate_data, results_gdf], ignore_index=True).sort_values(by=['Территория', '№ п/п'])

results_gdf


```
